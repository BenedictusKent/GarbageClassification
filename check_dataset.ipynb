```python
import os
import cv2
import numpy as np
from time import sleep
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
own = "img/own"
card = "cardboard"
glass = "glass"
metal = "metal"
paper = "paper"
plastic = "lastic"
trash = "trash"

LABELS = {card:0, glass:1, metal:2, paper:3, plastic:4, trash: 5}

axes = []
name = []
trial = []
bad_dataset = []

for label in LABELS:
    path = os.path.join(own, label)
    
    # read in improperly-labeled images
    if(len(os.listdir(path)) > 0):
        for image in os.listdir(label):
            try:
                if(label in image == False):
                    img = cv2.imread(path)
                    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                    img = cv2.resize(img, (512, 384))
                    trial.append(img)
                    name.append(image)
            except Exception as e:
                print(image)
    
    # all images properly labeled
    else:
        print("No images found in ", path)
    
    # show images
    cols = 6
    rows = int(len(trial) / cols) + 1
    figs = plt.figure(figsize=(17,15))
    for i in range(rows * cols):
        if(i < len(trial)):
            axes.append(figs.add_subplot(rows, cols, i+1))
            axes[-1].set_title(label)
            plt.imshow(trial[i])
    plt.show()
    
    # input bad datasets
    print("Put in bad dataset index(es) starting 0 or type in '"ok"' if finished")
    print("Inputted index(es) will delete the image automatically")
    while(True):
        temp = input("Input: ")
        if(temp != "ok"):
            bad_dataset.append(temp)
        else:
            break
            
    # clear output
    for i in range(len(bad_dataset)):
        string = name[bad_dataset[i]]
        if(os.path.exists(string)):
            os.remove(string)
    temp = input("Press any key to clear output...")
    axes.clear()
    name.clear()
    trial.clear()
    bad_dataset.clear()
    sleep(1)
    clear_output(wait=False)
```
