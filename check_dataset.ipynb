```python
import os
import cv2
import numpy as np
from time import sleep
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
own = "img/own"
card = "cardboard"
glass = "glass"
metal = "metal"
paper = "paper"
plastic = "plastic"
trash = "trash"

LABELS = {card:0, glass:1, metal:2, paper:3, plastic:4, trash: 5}

delete = 0
empty = 1
axes = []
name = []
trial = []
bad_dataset = []

for label in LABELS:
    path = os.path.join(own, label)
    print(path)
    
    # read in improperly-labeled images
    if(len(os.listdir(path)) > 0):
        for image in os.listdir(path):
            empty = 2
            impath = os.path.join(path, image)
            if((label in image) == False):
                empty = 0
                try:
                    img = cv2.imread(impath)
                    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                    img = cv2.resize(img, (512, 384))
                    trial.append(img)
                    name.append(image)
                except Exception as e:
                    print(image)
                  
    # all images properly labeled
    else:
        empty = 1
        print("No images found in ", path)
        
    # show images
    if(empty == 2):
        print("All images are all labelled properly")
    elif(empty == 0):
        cols = 5
        rows = int(len(trial) / cols) + 1
        figs = plt.figure(figsize=(17,15))
        for i in range(rows * cols):
            if(i < len(trial)):
                axes.append(figs.add_subplot(rows, cols, i+1))
                plt.imshow(trial[i])
        plt.show()
        
        # input bad datasets
        print("Put in bad dataset index(es) starting 0 or type in \"ok\" if finished")
        while(True):
            temp = input("Input: ")
            if(temp != "ok"):
                delete = 1
                bad_dataset.append(int(temp))
            else:
                break
                
        # clear output
        for i in range(len(bad_dataset)):
            print(name[bad_dataset[i]])
        
        if(delete == 1):
            temp = input("Type \"yes\" if you are sure to delete: ")
            if(temp == "yes"):
                for i in range(len(bad_dataset)):
                    string = name[bad_dataset[i]]
                    bpath = os.path.join(path, string)
                    if(os.path.exists(bpath)):
                        os.remove(bpath)
                        print(str(bpath), "removed")
                    
        temp = input("Press any key to clear output...")
        delete = 0
        axes.clear()
        name.clear()
        trial.clear()
        bad_dataset.clear()
        sleep(1)
        clear_output(wait=False)
```
