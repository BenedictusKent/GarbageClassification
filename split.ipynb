```python
import os
import cv2
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
def concat(image_path, pos):
    count = 0
    if len(os.listdir(image_path)) > 0:
        for image in os.listdir(image_path):
            try:
                path = os.path.join(image_path, image)
                img = cv2.imread(path, cv2.IMREAD_COLOR)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = cv2.resize(img, (512,384))
                if pos != 5:
                    dataset_trashless.append([img, pos])
                else:
                    trash.append([img, pos])
                count += 1
            except Exception as e:
                print(image)
    print(count)
```

```python
kaggle = "img/kaggle"
own = "img/own"

GLASS = "glass"
METAL = "metal"
PAPER = "paper"
PLASTIC = "plastic"
CARDBOARD = "cardboard"
TRASH = "trash"
LABELS = {CARDBOARD:0, GLASS:1, METAL:2, PAPER:3, PLASTIC:4, TRASH: 5}

trash = []
dataset_trashless = []
pos = 0

for label in LABELS:
    kpath = os.path.join(kaggle, label)
    opath = os.path.join(own, label)
    # append kaggle images to list
    print(kpath)
    concat(kpath, pos)
    # append own images to list
    print(opath)
    concat(opath, pos)
    # change pos
    pos += 1

print("Trashless total: ", str(len(dataset_trashless)))
print("Trash total: ", str(len(trash)))
```

```python
# randomize data
np.random.shuffle(dataset_trashless)
np.random.shuffle(trash)

# split variables 1.0
trashless_length = int(len(dataset_trashless) * 0.2)        
trash_length = int(len(trash) * 0.2)                        
train_length = len(dataset_trashless) - 2*trashless_length  
test_length = train_length + trashless_length               
val_length = test_length + trashless_length                 

# trashless split
train = dataset_trashless[:train_length]
test = dataset_trashless[train_length:test_length]
val = dataset_trashless[test_length:val_length]

# split variables 2.0
train_length = len(trash) - 2*trash_length
test_length = train_length + trash_length
val_length = test_length + trash_length

# trash split
train += trash[:train_length]
test += trash[train_length:test_length]
val += trash[test_length:]

# rerandomize
np.random.shuffle(train)
np.random.shuffle(test)
np.random.shuffle(val)

# split train test val
xtrain = []
ytrain = []
for features, label in train:
    xtrain.append(features)
    ytrain.append(label)
train.clear()

xtest = []
ytest = []
for features, label in test:
    xtest.append(features)
    ytest.append(label)
test.clear()

xval = []
yval = []
for features, label in val:
    xval.append(features)
    yval.append(label)
val.clear()

# change to numpy
xtrain = np.asarray(xtrain)
ytrain = np.asarray(ytrain)
xtest = np.asarray(xtest)
ytest = np.asarray(ytest)
xval = np.asarray(xval)
yval = np.asarray(yval)

print(xtrain.shape, ytrain.shape)
print(xtest.shape, ytest.shape)
print(xval.shape, yval.shape)
```

```python
import pickle

out = open("xtrain.pickle", "wb")
pickle.dump(xtrain, out)
out.close()

out = open("ytrain.pickle", "wb")
pickle.dump(ytrain, out)
out.close()

out = open("xtest.pickle", "wb")
pickle.dump(xtest, out)
out.close()

out = open("ytest.pickle", "wb")
pickle.dump(ytest, out)
out.close()

out = open("xval.pickle", "wb")
pickle.dump(xval, out)
out.close()

out = open("yval.pickle", "wb")
pickle.dump(yval, out)
out.close()
```
